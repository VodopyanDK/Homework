Нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. Перечень всех документов

```python
data_documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

2. Перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

```python
directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}
```

## Задание 1
Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру

Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится

Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам

Пункт 4. Пользователь по команде “as” может добавить полку в перечень

Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)


## Задание 2 (дополнительное)

Необходимо дополнить программу из задания 1 более продвинутыми командами.

Пункт 1. Пользователь по команде “ad” может добавить новый документ в данные

Пункт 2. Пользователь по команде “d” может удалить документ из данных

Пункт 3. Пользователь по команде “m” может переместить документ с полки на полку



```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
# вывод ключей
def cout_keys():
    return  ','.join([str(i) for i in list(directories.keys())]) 

# Пункт 1 
def define_owner(docs_number):   
    owner = str()
    for line in documents: 
        for keys, values in line.items(): 
            if (values == docs_number):
                owner = line['name']
    if not owner:
        return print('Документ не найден в базе') 
    else:
        return print('Владелец документа: ', owner) 
# Пункт 2
def define_shelf(shelf_number): 
    for key, vls in directories.items(): 
        if (shelf_number in vls ):
            return key
    return None
# Пункт 3
def full_info():
    for dictin in documents:
        shelf_count = define_shelf(dictin['number']) 
        # словарь для вывода
        new_dicti= {'№:': dictin['number'] ,'тип:' : dictin['type'],'владелец:': dictin['name'], 'полка хранения:': shelf_count }
        print(','.join( str(' ' + k +' ' + v) for k, v in new_dicti.items() ))
    return            
# Пункт 4
def adds_shelf(new_shelf): 
    if new_shelf in directories.keys():
        print('Такая полка уже существует. Текущий перечень полок:',cout_keys(),'.' )     
        return
    else:
        directories[new_shelf] = []
        print('Полка добавлена. Текущий перечень полок:', cout_keys(),'.' )     
        return
# Пункт 5
def del_shelf(d_shelf):
    if d_shelf in directories.keys():
        if (directories.get(d_shelf) == []): 
            del(directories[d_shelf])
            print('Полка удалена. Текущий перечень полок: ', ','.join([str(i) for i in list(directories.keys())]),'.' )
            return
        elif (directories.get(d_shelf) != []): 
            return print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', cout_keys(),'.' )
    else:
        return print('Такой полки не существует. Текущий перечень полок: ',cout_keys(),'.' )
# Дополнительное задание
# Пункт 1
def add_document(): 
    number = input('Введите номер документа: ')
    doc_type = input('Введите тип документа: ')
    name_owner = input('Введите владельца документа: ')
    new_shelf_number = input('Введите полку для хранения: ')
    doc_dictinary = {'type': doc_type, 'number': number, 'name': name_owner } 
    if new_shelf_number not in directories.keys():
        print('Такой полки не существует. Добавьте полку командой as.')
        print('Текущий список документов:')
        full_info()
        return
    else:
        documents.append(doc_dictinary)
        directories.get(new_shelf_number).append(number)
        print('Документ добавлен. Текущий список документов:')
        full_info() 
        return
# Пункт 2
def del_document():
    nmbr_docr = input('Введите номер документа:')
    for line in documents:
        for vals in line.values():
            if (vals == nmbr_docr): 
                documents.remove(line)
                print('Документ удален.')
                print('Текущий список документов:')
                full_info()
                return
    print('Документ не найден в базе.')
    print('Текущий список документов:')
    full_info()
    return

# Пункт 3
def moving_docs():
    number_d = input('Введите номер документа: ')
    number_shelf = input('Введите номер полки: ')
    if number_shelf in directories.keys():
        if (define_shelf(number_d) == None): 
            print('Документ не найден в базе:')
            print('Текущий список документов:')
            full_info()
            return
        else: 
            directories[define_shelf(number_d)].remove(number_d)
            directories[number_shelf].append(number_d) 
            print('Документ перемещен.')
            print('Текущий список документов:')
            full_info()
            return 
    else:
        return print('Такой полки не существует. Текущий перечень полок: ' + cout_keys()) 
            

def main():
    choice = input('Введите команду: ')
    if (choice == 'p'):
        docs_number = input('Введите номер документа: ')
        define_owner(docs_number)
    elif (choice == 's'):
        shelf_number = input('Введите номер документа')
        owner = define_shelf(shelf_number)
        if not owner :
            print('Документ не найден в базе')
        else:
            print('Документ хранится на полке: ', owner)
    elif (choice == 'l'):
        full_info()
    elif (choice == 'as'):
        n_shelf = input('Введите номер полки: ')
        adds_shelf(n_shelf)
    elif (choice == 'ds'):
        number_of_shelf = input('Введите номер полки: ')
        del_shelf(number_of_shelf)
    elif (choice == 'ad'):
        add_document()
    elif (choice =='d'):
        del_document()
    elif (choice == 'm'):
        moving_docs()
    else:
        print('Неизвестная команда.')
```

```python
main()
```
